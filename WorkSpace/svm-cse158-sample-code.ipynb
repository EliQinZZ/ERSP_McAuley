```python
import numpy
import urllib.request
import scipy.optimize
import random
from sklearn import svm

def parseData(fname):
  for l in urllib.request.urlopen(fname):
    yield eval(l)

print ("Reading data...")
data = list(parseData("http://jmcauley.ucsd.edu/cse255/data/amazon/book_images_5000.json"))
print ("done")
```

```python
X = [b['image_feature'] for b in data]
y = ["Children's Books" in b['categories'] for b in data]

X_train = X[:2800]
y_train = y[:2800]

X_valid = X[2800:4000]
y_valid = y[2800:4000]

X_test = X[4000:]
y_test = y[4000:]
```

```python
cList = []
for i in range(-3, 33):
    cList.append(2 ** i)
print(cList)
```

```python
train_errors = []
valid_errors = []
for c in cList:
    clf = svm.SVC(C = c)
    clf.fit(X_train, y_train)
    
    train_predictions = clf.predict(X_train)
    valic_predictions = clf.predict(X_valid)
```

```python
# Create a support vector classifier object, with regularization parameter C = 1000
clf = svm.SVC(C=1000)
clf.fit(X_train, y_train)

train_predictions = clf.predict(X_train)
test_predictions = clf.predict(X_test)
```

```python
TP = 0
TN = 0
FP = 0
FN = 0
for i in range(len(train_predictions)):
    test = y_test[i]
    pred = test_predictions[i]
    
    if(test == True and pred == True):
        TP += 1
    if(test == False and pred == False):
        TN += 1
    if(test == True and pred == False):
        FN += 1
    if(test == False and pred == True):
        FP += 1
    
    print("{}\t{}".format(y_test[i], test_predictions[i]))
```
