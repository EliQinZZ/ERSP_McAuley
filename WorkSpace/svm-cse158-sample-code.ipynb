```python
import numpy
import urllib.request
import scipy.optimize
import random
from sklearn import svm
import gzip

#def parseData(fname):
#  for l in urllib.request.urlopen(fname):
#    yield eval(l)

print ("Reading data...")
tin = gzip.open('../RunningBikingTrainValidTest/training_set_running_and_biking.json.gz', 'rb')
vin = gzip.open('../RunningBikingTrainValidTest/validation_set_running_and_biking.json.gz', 'rb')
train_list = []
valid_list = []
for l in tin:
    l = l.decode('ascii')
    dic = eval(l)
    train_list.append(dic)
for l in vin:
    l = l.decode('ascii')
    dic = eval(l)
    valid_list.append(dic)
print ("done")
```

```python
import numpy
def generate_features(dataList):
    feature_list = []
    for data in dataList:
        curr_features = []
        curr_features.append(numpy.amin(data))
        curr_features.append(numpy.mean(data))
        curr_features.append(numpy.amax(data))
        
        feature_list.append(curr_features)
    return feature_list
```

```python
X_train_list = generate_features([b['speed'] for b in train_list])
y_train_list = [b['sport'] is 'run' for b in train_list]

X_valid_list = generate_features([b['speed'] for b in valid_list])
y_valid_list = [b['sport'] is 'run' for b in train_list]
```

```python
cList = []
for i in range(-3, 33):
    cList.append(2 ** i)
print(cList)
```

```python
def findErr(pred, real):
    total = 0
    errors = 0
    for i in range(len(pred)):
        total += 1
        if pred[i] != real[i]:
            errors += 1
    
    return errors / total
```

```python
train_errors = []
valid_errors = []
for c in cList:
    print("Running for c = " + str(c))
    clf = svm.SVC(C = c)
    clf.fit(X_train_list, y_train_list)
    
    train_predictions = clf.predict(X_train_list)
    valid_predictions = clf.predict(X_valid_list)
    
    train_errors.append(findErr(train_predictions, y_train_list))
    valid_errors.append(findErr(valid_predictions, y_valid_list))
    
    print("{}\t{:.2}\t{:.3}".format(c, train_errors[len(train_errors) - 1], valid_errors[len(valid_errors) - 1]))
```

```python
for i in range(len(train_errors)):
    print("{}\t{:.2}\t{:.3}".format(cList[i], train_errors[i], valid_errors[i]))
```
