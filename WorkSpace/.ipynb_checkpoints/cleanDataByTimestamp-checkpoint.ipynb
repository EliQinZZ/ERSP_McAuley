```python
import gzip
```

```python
def cleanDataByTimestamp(inFile, outFile, threshold):
    entryKept = 0
    entryDiscarded = 0
    f = open('RESULT_cleanDataByTimestamp', 'a')
    for l in inFile:
        decodedl = l.decode('ascii')
        dic = eval(decodedl)
        
        numOfTimestamp = len(dic['timestamp'])
        
        if(numOfTimestamp < threshold):
            entryDiscarded += 1
            f.write("Record discarded with {}\n".format(numOfTimestamp))
            continue
            
        outFile.write(l)
        entryKept += 1
        f.write("Record kept with {}\n".format(numOfTimestamp))
        
    print('{}/{}'.format(entryKept, entryKept + entryDiscarded))
    f.close()
```

```python
zin = gzip.open('../Data/workoutsWithSpeed_run.json.gz', 'r')
zout = gzip.open('cleanedTimestamp.json.gz', 'ab')
cleanDataByTimestamp(zin, zout, 100)
zin.close()
zout.close()
```
